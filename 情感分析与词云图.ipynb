```python
from wordcloud import WordCloud, ImageColorGenerator
import matplotlib.pyplot as plt
import pandas as pd
import jieba
```

```python
df = pd.read_csv('liulang.csv', header=None)
```

```python
df.head()
```

```python
text = ''
for line in df[6][1:]:
    text += ' '.join(jieba.cut(line, cut_all=False))
```

```python
filefath = 'stopwords.txt'
stopwords = [line.strip() for line in open(filefath,'r').readlines()]
final = ''
fn2 = open("new.txt", "a+", encoding='utf-8')
for word in text:
    if word not in stopwords:
        if word != '\t':
           final +=word
           final +=" "
fn2.write(final)   #写入去掉停用词的内容
```

```python
from snownlp import SnowNLP
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
f=open('new.txt','r',encoding='UTF-8')
list=f.readlines()
sentimentslist=[]
for i in list:
    s=SnowNLP(i)
    sentimentslist.append(s.sentiments)
plt.hist(sentimentslist,bins=np.arange(0,1,0.01),facecolor='g')
plt.xlabel('Sentiments Probability')
plt.ylabel('Quantity')
plt.title('Analysis of Sentiments')
plt.show()
```

```python
filefath = 'stopwords.txt'
stopwords = [line.strip() for line in open(filefath,'r').readlines()]
final = ''
fn2 = open("newword.txt", "a+", encoding='utf-8')
for word in df[6][1:]:
    if word not in stopwords:
        if word != '\t':
           final +=word
           final +=" "
fn2.write(final)   #写入去掉停用词的内容

text = ''
for line in open(r'newword.txt','r',encoding='utf-8').readlines():
    text += ' '.join(jieba.cut(line, cut_all=False))

backgroud_Image = plt.imread('liulang.jpg') #指定词云的形状引用imread()函数
print('加载图片成功！')
'''设置词云样式'''
wc = WordCloud(
    background_color='white',  #指定图片的背景颜色
    mask=backgroud_Image,    #指定词云的形状
    font_path='C:\Windows\Fonts\simhei.ttf',
    max_words=3000,
    max_font_size=100,
    random_state=50, # 设置有多少种随机生成状态，即有多少种配色方案
)
wc.generate_from_text(text)
  #generate()向wordcloud中加载文本  
print('开始加载文本')
# 看看词频高的有哪些,把无用信息去除
process_word = WordCloud.process_text(wc, text) #将长文本分词并去除屏蔽词
sort = sorted(process_word.items(), key=lambda e:e[1], reverse=True)
print(sort[:50])
#改变字体颜色
img_colors = ImageColorGenerator(backgroud_Image)  #从背景图片生成颜色值
#字体颜色为背景图片的颜色
wc.recolor(color_func=img_colors) #recolor([random_state, color_func, colormap])  对现有输出重新着色。重新上色会比重新生成整个词云快很多。
# 显示词云图
plt.imshow(wc)
# 是否显示x轴、y轴下标
plt.axis('off')
wc.to_file("liulang2.jpg") #将词云输出的图像原件.png或.jpg给保存
print('生成词云成功!')
```
